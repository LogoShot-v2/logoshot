```python
def excelhandle():
    os.chdir('/service/trademark/labels/processed')
    path = os.getcwd()
    #print(path)
    dirs = sorted(os.listdir())
    catecounts = []
    
    with open('/home/jack/4_Selenium/ALL.csv','a+' ,newline='',encoding="utf-8") as fd:
        writer = csv.writer(fd)
        writer.writerow(['caseno','category'])
    
    for x in dirs:
        cate = x.split('.')[0]
        df = pd.read_csv(x)
        casenos = df['caseno'].to_list()
        print("Category: %s, Count: %d"%(cate,len(casenos)))
        catecounts.append((cate,len(casenos)))
        
        with open('/home/jack/4_Selenium/Data.txt','a+',encoding='utf-8') as f:
            f.write("Category: %s, Count: %d\n"%(cate,len(casenos)))
        
        with open('/home/jack/4_Selenium/ALL.csv','a+' ,newline='',encoding="utf-8") as fd:
            writer = csv.writer(fd)
            #writer.writerow(['caseno','category'])
            for i in range(len(casenos)):
                if casenos[i] == 'nan':
                    continue
                else:
                    writer.writerow([casenos[i],cate])
#     shutil.copy("/home/jack/4_Selenium/ALL.csv", "/service/trademark/labels/ALL.csv")
    return catecounts
```

```python
def findmaxn(n,l):
    cnt = 0
    m = []
    while cnt < n:
        m.append(max(l))
        l = list(set(l) - set([max(l)]))
        cnt += 1
    return m
def findminn(n,l):
    cnt = 0
    m = []
    while cnt < n:
        m.append(min(l))
        l = list(set(l) - set([min(l)]))
        cnt += 1
    return m
```
