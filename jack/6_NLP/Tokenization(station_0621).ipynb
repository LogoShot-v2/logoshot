```python
def find_chinese(file):
    pattern = re.compile(r'[^\u4e00-\u9fa5]')
    chinese = re.sub(pattern, '', file)
    return chinese
def find_unchinese(file):
    pattern = re.compile(r'[\u4e00-\u9fa5]')
    unchinese = re.sub(pattern,"",file)
    return unchinese
```

```python
%%time
#Main
# (1) Get all casenos
cur.execute("SELECT caseno FROM trademark")
casenos = cur.fetchall()
print(len(casenos))

# (2) For each caseno, find all text
# (3) Find Chinese Only
# (4) Monpa cut, stopword removal!!!
D = {}
cnt = 0
monpa.load_userdict("userdict.txt")
f = open("Rs/Result.txt","a+")


for cno in casenos:
#     print(cno)
#     rid = random.randint(0,len(casenos))
#     rid = cnt
    
#     print(casenos[rid][0])
    text = getalltext(cno[0])
#     getall(casenos[rid][0])
#     print(text)
    conly = find_chinese(text)
    
    if len(conly) <= 200:
        tokens = monpa.cut(conly)
    else:
        t = []
        tokens = []
        part = 0
        while (200*part) < len(conly):
            t.append(monpa.cut(conly[part*200:(part+1)*200]))
            part += 1
        t.append(monpa.cut(conly[part*200:]))

        for i in t:
            tokens += i

    for i in tokens:
        if i in "以及之或與和的為":
            tokens.remove(i)
#     print(".....")
#     print(tokens)
    D[cno[0]] = tokens

    f.write(str(cno[0])+",")
    for i in tokens:
        f.write(i)
        f.write(" ")
    f.write("\n")
    
    cnt += 1
    if (cnt % 10000 == 0):
        print("Progress : %d / 656459"%cnt)
        with open("Progress.txt","a+") as p:
            p.write("Progress : %d / 656459\n"%cnt)
        f.close()
        f = open("Rs/Result%s.txt"%str(cnt//10000),"a+")
#     if cnt == 50:
#         break
#     print("~~~~~")
# print(D)

f.close()
print("Done")
with open("Progress.txt","a+") as p:
    p.write("Done")

# (5) Save as dict, txt
# with open("Dict.pk","wb") as f:
#     pk.dump(D,f)

print(len(D.keys()))
```

```python
def getalltext(ID):
    
    text = ""
    cur.execute("SELECT trademark_name FROM trademark WHERE caseno = %s"%ID)
    r = cur.fetchall()
    if (r[0][0] != None):
        text += r[0][0]
    
    cur.execute("SELECT goods_denomination FROM rcc WHERE caseno = %s"%ID)
    r = cur.fetchall()
    if (r != []):
        for i in r:
            if (i[0] != None):
                text += i[0]

    cur.execute("SELECT achinese,address FROM rco WHERE caseno = %s"%ID)
    r = cur.fetchall()
    if (r != []):
        for i in r:
    #         print(i[0])
            if (i[0] != None):
                ii = i[0].split("\u3000")
                for j in ii:
                    if len(j) != 3:
                        text += j
            if (i[1] != None):
                text += i[1]

    cur.execute("SELECT chinese FROM text WHERE caseno = %s"%ID)
    r = cur.fetchall()
    if (r != []):
        if (r[0][0] != None):
            text += r[0][0]
    
    return text
```
