# 合併100年前後資料
100年第17卷 ~ 101年第12卷的RegContent.xml有36個欄位，但101年第13卷 ~ 110年第5卷的RegContent.xml只有25個欄位，前者比後者多11個欄位，  
差異的欄位為：
Design_Chinese / Design_English / Design_Japanese / Design_Numeral / Page_Chnum / Page_Num / Published / RegContentAgent / RegContentSN / RegContentClassification / RegContentSN / RegContentOwner / RegContentSN / RegContentPicture / RegContentSN  
目前處理方式為將多出的11個欄位刪除

```python
import pandas as pd
import re
from datetime import datetime
```

```python
#民國轉西元
def to_common_era(x):
    date = x.split('.')
    date[0] = str(int(date[0]) + 1911)
    date = '-'.join(date)
    return datetime.strptime(date, "%Y-%m-%d")

def merger_old_new(old, new):
    col_to_be_delete = 'Design_Chinese / Design_English / Design_Japanese / Design_Numeral / Page_Chnum / Page_Num / Published / RegContentAgent / RegContentSN / RegContentClassification / RegContentSN / RegContentOwner / RegContentSN / RegContentPicture / RegContentSN'
    col_to_be_delete = set(col_to_be_delete.split(' / '))
    cols = set(old.columns).intersection(col_to_be_delete)
    old.drop(columns=cols, inplace=True);
    return pd.concat([old, new], axis=0, ignore_index=True)

def save(df, fname):
    df.to_csv(path+fname+'.csv', index=False, encoding="utf-8")
```

### '\u3000'是全形空白，有些空值是用全形空白表示的，替換程空字串

```python
RCC = merger_old_new(RCC_old, RCC_new)
RCC['Class'] = RCC['Class'].apply(lambda x: x.replace('\u3000', ''))
```
