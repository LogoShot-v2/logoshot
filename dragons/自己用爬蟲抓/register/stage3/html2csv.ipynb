```python
import pandas as pd
import os
import html
from bs4 import BeautifulSoup
import csv
import numpy as np
```

```python
def get_content(html):
    html_text = html.read()
    soup = BeautifulSoup(html_text, 'html.parser')
    td = soup.find_all('td')
    
    flag = 0
    text_list = []
    for ele in td:
        text = ele.text.strip()
        if text in ["申請案號", "圖樣中文", "圖樣英文", "圖樣日文", "圖樣記號"]:
            flag = 1
#             print(text)
            continue
        if(flag == 1):
#             print(text)
            flag = 0
            text_list.append(text)
    return text_list
```

```python
# for dir_ in ["output_wen", "output_dragons"]:
for dir_ in ["output_dragons2"]:    
    file_names = os.listdir(dir_)
    for i in range(len(file_names)):
        file_name = file_names[i]
        file_path = os.path.join(dir_, file_name)
        with open(file_path, 'r', encoding="utf-8") as f:
            text_list = get_content(f)
            with open('./register.csv','a', newline='') as fd:
                writer = csv.writer(fd)
                writer.writerow(text_list)
```
