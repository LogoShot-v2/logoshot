```python
class myThread(threading.Thread): #继承threading.Thread类
    def __init__(self,name,link_range):
        threading.Thread.__init__(self)
        self.name = name
        self.link_range = link_range
    def run(self):  #重写成员方法
        print ("Starting " + self.name)
        crawler(self.name, self.link_range)
        print ("Exiting " + self.name)

def crawler(threadName, link_range):
    CaseNos = CaseNo_list[link_range[0]: link_range[1]]
    print(threadName, link_range)
    fail_num = 0
    driver = webdriver.Chrome('./chromedriver')    
    for i in range(0, len(CaseNos),10):
        print("{} Progress: {}/{}, Fail number: {}".format(threadName, i, len(CaseNos), fail_num))
        try:  
            driver.get("https://twtmsearch.tipo.gov.tw/OS0/OS0201.jsp?l6=zh_TW&isReadBulletinen_US=&isReadBulletinzh_TW=true")
            time.sleep(5)
            element = driver.find_element_by_id('applNo')
            element.send_keys(str(CaseNos[i]))
            for j in range(9):
                button = driver.find_element_by_xpath("//*[@id='form1']/fieldset/div[4]/div/table/tbody/tr/td[2]/img")
                button.click()
                time.sleep(0.01)
                temp = ('applNo'+"%s" % int(j+1))
                element = driver.find_element_by_id(temp)
                element.send_keys(str(CaseNos[i+j+1]))
            button = driver.find_element_by_id("querySubmit")
            button.click()
            time.sleep(5)

            alist = []
            blist = [] #caseNo
            for link in driver.find_elements_by_xpath("//*[@id='APPL_NO']/a"):
                #print("link",link)
                key = link.get_attribute('href')[29:70]
                CaseNo_find = link.text
                if(key not in alist and CaseNo_find not in blist):
                    alist.append(key)
                    blist.append(CaseNo_find)
        except:
            print("11111111111")
            fail_num += 10
            continue
        if(len(alist) == 10 and len(blist) == 10):
            with open('./key_old({}).csv'.format(threadName),'a', newline='') as fd:
                writer = csv.writer(fd)
                for k in range(10):
                    writer.writerow([blist[k],alist[k]])
        else:
            print("2222222222222")
            fail_num += 10
            continue
```

```python
df1 = pd.read_csv('trademark_old.csv')
df2 = pd.read_csv('key_old.csv')
CaseNo_list1 = df1['CaseNo'].tolist()
CaseNo_list2 = df2['CaseNo'].tolist()

CaseNo_list = list(set(CaseNo_list1)-set(CaseNo_list2))
CaseNo_list.sort()

link_range = []
sep = len(CaseNo_list)//thread_num
for i in range(thread_num):
    if(i != thread_num-1):
        link_range.append((sep*i, sep*(i+1)))
    else:
        link_range.append((sep*i, len(CaseNo_list)))

thread_list=[]

#创立新线程
for i in range(1,thread_num+1):
    thread=myThread('Thread-'+str(i),link_range[i-1])
    thread.start() #线程开始执行
    thread_list.append(thread)

#等待所有线程完成
for t in thread_list:
    t.join()
```
