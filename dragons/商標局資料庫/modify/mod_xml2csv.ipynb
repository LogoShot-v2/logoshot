```python
import xml.etree.ElementTree as ET
import pandas as pd
import os
```

```python
all_case = []
dossiers = os.listdir("./raw")
dossiers.sort()
for index, dossier in enumerate(dossiers): 
    if(dossier == ".DS_Store"):
        continue    
    path = os.path.join("raw", dossier)
    sub_paths = [os.path.join(path, file) for file in os.listdir(path)]
    sub_paths.sort()
    for sub_path in sub_paths:
        if(sub_path[-4:] != ".xml" or sub_path[-9:] == "index.xml"):
            continue
        tree = ET.parse(sub_path)
        root = tree.getroot()
        mod_type  = sub_path.split("/")[-1].split(".")[0]
        for case in root:
            case_dict = {"doc":dossier, "mod_type":mod_type} 
            for tag in case:
                case_dict[tag.tag] = tag.text
#                 if("#-sep-#" in str(tag.text)):
#                     print("the sep notation is bad QQ")
                for sub_tag in tag:
                    name = tag.tag+'#'+sub_tag.tag
                    if tag.tag in case_dict.keys():
                        del case_dict[tag.tag]
                    
#                     if("#-sep-#" in str(sub_tag.text)):
#                         print("the sep notation is bad QQ")
                    
                    if name not in case_dict.keys():
                        case_dict[name] = [str(sub_tag.text)]
                    else:
                        case_dict[name].append(str(sub_tag.text))
            all_case.append(case_dict)    
    if((index+1) % 10 == 0):
        print("The progress is {} / {}".format(index+1, len(dossiers)-1)) # 減去.DS_Store檔
```

```python
# pd.DataFrame([{"x":1, "y":2}, {"y":3, "z":4}])
df = pd.DataFrame(all_case)
pd.set_option('display.max_columns', None)
df.head()
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: "#-sep-#".join(x) if type(x) == list else x)
```

```python
df[df['mod_type']=="Alteration"].iloc[2:3,:].dropna(axis = 1)
```

```python
df[df['RegisterNo'].isna()]
df = df.drop(index=0)
```

```python
df.to_csv("modify.csv", index = False)
```
